```python
# Fonction permettant de scorer les indicateurs d'un data set pour un indicateur et a une date donnée
# pour tracer le top 10 de chaque indicateur
def scoring_indicator(frame,indicateur, date):
    
#création d'un sous data frame ne contenant que l'indicateur voulu    
    frame_work=frame.copy()
    frame_work['bool'] = frame_work['Indicator Name'].apply(lambda indic_name: indicateur in indic_name) 
    masque_scor=frame_work['bool'] == True
    frame_indicator=frame_work[masque_scor]
    frame_indicator=frame_indicator.drop(['bool'],axis=1)
    
# normaliser la valeur des pays sur la valeur max parmis tout les pays  
    frame_normal = frame_indicator.loc[:,'2000':'2015'].apply(lambda x: x/x.max()*100, axis=0)   

#selectionner la date      
    frame_date=frame_normal.loc[:,date]
    frame_date=frame_date.sort_values(ascending=False)

#afficher le top 10
    frame_top10=pd.DataFrame(frame_date.iloc[0:10])
    frame_top10['pays']=frame_top10.index
    print(frame_top10)

#plot du Top 10       
    x=frame_top10['pays']
    y=frame_top10['2015']
    plt.plot(x,y,'o')
    plt.xticks(rotation=90)
  

 
```

```python
# Fonction permettant de scorer les indicateurs d'un data set pour une liste d'indicateur et a une date donnée
# pour tracer le top 10 de chaque indicateur
def scoring_indicator_list(frame,liste_indicateur, liste_date):
    
    frame_work=frame.copy()
    for indicateur in liste_indicateur:
        
#création d'un sous data frame ne contenant que l'indicateur voulu        
        frame_work['bool'] = frame_work['Indicator Name'].apply(lambda indic_name: indicateur in indic_name) 
        masque_scor=frame_work['bool'] == True
        frame_indicator=frame_work[masque_scor]
        frame_indicator=frame_indicator.drop(['bool'],axis=1)
    
# normaliser la valeur des pays sur la valeur max parmis tout les pays  
        frame_normal = frame_indicator.loc[:,'2000':'2015'].apply(lambda x: x/x.max()*100, axis=0)   

        for date in liste_date:
            
#selectionner la date      
            frame_date=frame_normal.loc[:,date]
            frame_date=frame_date.sort_values(ascending=False)

#afficher le top 10
            frame_top10=pd.DataFrame(frame_date.iloc[0:10])
            frame_top10['pays']=frame_top10.index
            frame_top10=frame_top10.reset_index(drop=True)
            print('Pour l\'année',date,'les pays en top 10 pour l\'indicateur :',indicateur )
            print('-'*20)
            print(frame_top10)

#plot du Top 10       
            x=frame_top10['pays']
            y=frame_top10[date]
            plt.plot(x,y,'o')
            plt.xticks(rotation=90)
            plt.show()
  
```

```python
# Fonction permettant de scorer les pays d'un data set pour une liste d'indicateur et a une date donnée

def scoring_total (frame,list_indicateur , date):
    frame_work=frame.copy()
    frame_total=pd.DataFrame()
    
        
    for indicateur in list_indicateur:
        
        #création d'un sous data frame ne contenant que l'indicateur voulu        
        frame_work['bool'] = frame_work['Indicator Name'].apply(lambda indic_name: indicateur in indic_name ) 
        masque_scor=frame_work['bool'] == True
        frame_indicator=frame_work[masque_scor]
        frame_indicator=frame_indicator.drop(['bool'],axis=1)

        # normaliser la valeur des pays sur la valeur max parmis tout les pays 

        frame_normal = frame_indicator.loc[:,'2000':'2015'].apply(lambda x: x/x.max(), axis=0)  
        frame_normal=frame_normal.reset_index()
        frame_total[indicateur]=frame_normal.loc[:,date]


    #donne le score totale normaliser pour chaque pays
    frame_total['score total']= frame_total.sum(axis=1)     
    frame_total['score norm']=frame_total['score total']/frame_total['score total'].max()*100
    frame_total=frame_total.drop('score total',axis=1)
    frame_total=frame_total.set_index(final_data.index.unique())
    
    frame_total=frame_total.sort_values(by= ['score norm'], ascending=False)  
    
        
    #afficher le top 10
    frame_top10=pd.DataFrame(frame_total['score norm'].iloc[0:10])
    frame_top10['pays']=frame_top10.index
    frame_top10=frame_top10.reset_index(drop=True)
    print('Pour l\'année',date,'les pays en top 10 sont :')
    print('-'*20)
    print(frame_top10)

    #plot du Top 10       
    x=frame_top10['pays']
    y=frame_top10['score norm']
    plt.plot(x,y,'o')
    plt.xticks(rotation=90)
    plt.show()
     
        
    print('Pour l\'année',date) 
 
    return(frame_total)     


```

```python
# Fonction permettant tracer l'evolution d'un indicateurs d'un data set pour le top 9
def evolution_indicateur(frame,indicateur):
    
    frame_work=frame.copy()
            
    #création d'un sous data frame ne contenant que l'indicateur voulu        
    frame_work['bool'] = frame_work['Indicator Name'].apply(lambda indic_name: indicateur in indic_name) 
    masque_scor=frame_work['bool'] == True
    frame_indicator=frame_work[masque_scor]
    frame_indicator=frame_indicator.drop(['bool'],axis=1)
    
    #tri par ordre croissant selon l'année 2015 pour selectionner les pays qui possède la plus grande valeur en 2015
    frame_indicator=frame_indicator.sort_values(by=['2015'],ascending=False)
    
    #afficher le top 9
    frame_top10=pd.DataFrame(frame_indicator.iloc[0:9])
    frame_top10=frame_top10.drop(['Indicator Name'],axis=1)
    frame_top10=frame_top10.drop(['Country Name'],axis=1)
    print('Evolution entre 2000 et 2015 des 9 meilleirs pays pour l\'indicateur :',indicateur )
    
    #diviser les valeur par 1000 ==> plus clair
    frame_top10=frame_top10/1000
    
## plot 

    #attribution des x et y dans un dataframe
    df=pd.DataFrame({'x': range(2000,2016), 'y1': frame_top10.iloc[0], 'y2': frame_top10.iloc[1], 
                 'y3': frame_top10.iloc[2], 'y4': frame_top10.iloc[3],
                 'y5': frame_top10.iloc[4], 
                 'y6': frame_top10.iloc[5], 'y7': frame_top10.iloc[6], 
                 'y8': frame_top10.iloc[7], 'y9': frame_top10.iloc[8] })
    
    #donne le noms des pays en top 9 aux colonnes
    myList = list(df.columns)
    
    for i in range(0,9):
        
        myList[i+1] = frame_top10.T.columns[i]
    
    df.columns = myList
    
    # Initialise la figure
    plt.style.use('seaborn-darkgrid')
    # couleur 
    palette = plt.get_cmap('Set1')
    # boucle pour tracé les différentes courbes 
    num=0
    for column in df.drop('x', axis=1):
        num+=1
        # position des graph
        plt.subplot(3,3,num)
        # plot chaques autres courbes
        for v in df.drop('x', axis=1):
            plt.plot(df['x'], df[v], marker='', color='grey', linewidth=0.6, alpha=0.3)
        # Plot la courbre principale
        plt.plot(df['x'], df[column], marker='', color=palette(num), linewidth=2.4, alpha=0.9, label=column)
        # fixe les bornes
        plt.xlim(2000,2015)
        plt.ylim(20,140)
        plt.xticks(rotation=30)
        # gère les ticks
        if num in range(7) :
            plt.tick_params(labelbottom=False)
        if num not in [1,4,7] :
            plt.tick_params(labelleft=False)
        # titre des graphes
        plt.title(column, loc='left', fontsize=12, fontweight=0, color=palette(num) )
    # titre general
    plt.suptitle('Evolution entre 2000 et 2015 des 9 meilleurs pays'  , fontsize=13, fontweight=0, color='black', style='italic', y=1.02)
    # titre des axes
    plt.text(1989, -40, 'Date', ha='center', va='center')
    plt.text(1959, 230, '{}'.format(indicateur), ha='center', va='center', rotation='vertical')
    

```

```python
# Fonction permettant tracer l'evolution VS de deux indicateurs d'un data set pour un pays donné
def scater_indicator(frame,indicateur1,indicateur2,pays):
    
    frame_work=frame.copy()
    #création d'un sous data frame ne contenant que les indicateurs du pays voulu    
    frame_work['bool'] = frame_work['Country Name'].apply(lambda pays_name: pays in pays_name) 
    masque_scor=frame_work['bool'] == True
    frame_pays=frame_work[masque_scor]
    frame_pays=frame_pays.drop(['bool'],axis=1)
    
        
    #création d'un sous data frame ne contenant que l'indicateur1 voulu    
    frame_pays['bool'] = frame_pays['Indicator Name'].apply(lambda indic_name: indicateur1 in indic_name) 
    masque_scor=frame_pays['bool'] == True
    frame_pays_indicator1=frame_pays[masque_scor]
    frame_pays_indicator1=frame_pays_indicator1.drop(['bool'],axis=1)
    frame_pays_indicator1=frame_pays_indicator1.drop(['Country Name'],axis=1)
    frame_pays_indicator1=frame_pays_indicator1.drop(['Indicator Name'],axis=1)
    frame_pays=frame_pays.drop(['bool'],axis=1)
    
    
    #création d'un sous data frame ne contenant que l'indicateur2 voulu    
    frame_pays['bool'] = frame_pays['Indicator Name'].apply(lambda indic_name: indicateur2 in indic_name) 
    masque_scor=frame_pays['bool'] == True
    frame_pays_indicator2=frame_pays[masque_scor]
    frame_pays_indicator2=frame_pays_indicator2.drop(['bool'],axis=1)
    frame_pays_indicator2=frame_pays_indicator2.drop(['Country Name'],axis=1)
    frame_pays_indicator2=frame_pays_indicator2.drop(['Indicator Name'],axis=1)
    
    #frame de travail
    df=pd.DataFrame({indicateur1: frame_pays_indicator1.loc[pays], indicateur2: frame_pays_indicator2.loc[pays]})
     
    #scaterplot   
    x=df.iloc[:,0]
    y=df.iloc[:,1]
    plt.plot(x,y,'o',alpha=0.5)
    plt.xlabel(indicateur1)
    plt.ylabel(indicateur2)
    title1 = 'Evolution entre 2000 et 2015 de \n {}'.format(indicateur1)
    title2 ='\n par rapport à \n {}'.format(indicateur2)
    plt.title(title1+title2)
    
```

```python
# Fonction permettant de scorer les pays d'un data set en fonction d'une liste d'indicateur et à une date donnée
# et trace les graphes radars du Top10

def radar_top10 (frame,list_indicateur , date):
    
    frame_work=frame.copy()
    frame_total=pd.DataFrame()
    
        
    for indicateur in list_indicateur:
        
        #création d'un sous data frame ne contenant que l'indicateur voulu        
        frame_work['bool'] = frame_work['Indicator Name'].apply(lambda indic_name: indicateur in indic_name ) 
        masque_scor=frame_work['bool'] == True
        frame_indicator=frame_work[masque_scor]
        frame_indicator=frame_indicator.drop(['bool'],axis=1)

        # normaliser la valeur des pays sur la valeur max parmis tout les pays 

        frame_normal = frame_indicator.loc[:,'2000':'2015'].apply(lambda x: x/x.max()*100, axis=0)  
        frame_normal=frame_normal.reset_index()
        frame_total[indicateur]=frame_normal.loc[:,date]


    #donne le score totale normaliser pour chaque pays
    frame_total['score total']= frame_total.sum(axis=1)     
    frame_total['score norm']=frame_total['score total']/frame_total['score total'].max()*100
    #frame_total=frame_total.drop('score total',axis=1)
    frame_total=frame_total.set_index(final_data.index.unique())
    
    frame_total=frame_total.sort_values(by= ['score norm'], ascending=False)  
    
        
    #afficher le top 10
    frame_top10=frame_total.iloc[0:10,0:8]
    frame_top10['Country Name']=frame_top10.index
    print('Pour l\'année',date,'les pays en top 10 sont :')
    print('-'*20)
    
    #Radar plot pour chaque pays
    for i in range(len(frame_top10.iloc[0:10])):
        df=pd.DataFrame()
        df['Valeurs']=frame_top10.iloc[i]
        df['Indic']=frame_top10.columns
        fig = px.line_polar(df, r='Valeurs', theta='Indic', line_close=True)
        fig.update_traces(fill='toself')
        print(frame_top10.iloc[i,8])
        print ('pays numéro :{}'.format(i+1))
        fig.show()
        
    #Affiche le frame du top 10
    frame_top10=frame_top10.drop('Country Name', axis=1)
    return(frame_top10)     

```

```python
# Fonction permettant de scorer les pays d'un data set en fonction d'une liste d'indicateur et à une date donnée
def bar_race (frame,list_indicateur , date):
    
    frame_work=frame.copy()
    frame_total=pd.DataFrame()
    frame_evol=pd.DataFrame()
    
    #for date in list_date :  
        
    for indicateur in list_indicateur:

        #création d'un sous data frame ne contenant que l'indicateur voulu        
        frame_work['bool'] = frame_work['Indicator Name'].apply(lambda indic_name: indicateur in indic_name ) 
        masque_scor=frame_work['bool'] == True
        frame_indicator=frame_work[masque_scor]
        frame_indicator=frame_indicator.drop(['bool'],axis=1)

        # normaliser la valeur des pays sur la valeur max parmis tout les pays 

        frame_normal = frame_indicator.loc[:,'2000':'2015'].apply(lambda x: x/x.max(), axis=0)  
        frame_normal=frame_normal.reset_index()

        #for date1 in list_date1 : 

        frame_total[indicateur]=frame_normal.loc[:,date]


    #donne le score totale normaliser pour chaque pays
    frame_total['score total']= frame_total.sum(axis=1)     
    frame_total['score norm']=frame_total['score total']/frame_total['score total'].max()*100
    frame_total=frame_total.drop('score total',axis=1)
    frame_total=frame_total.set_index(frame_work.index.unique())
    
        
    frame_evol[date]=frame_total['score norm']
    
    return(frame_evol)     
```

```python
# Fonction permettant de scorer les pays d'un data set en fonction d'une liste d'indicateur,
# d'une liste de date et de deux indicateurs à contraindre

def bar_race_medif (frame, list_indicateur, liste_date, indic_med1, indic_med2):
    
    frame_work=frame.copy()
    frame_total=pd.DataFrame()
    frame_total1=pd.DataFrame()
    frame_evol=pd.DataFrame()
    frame_total_med=pd.DataFrame()
    
    for date in liste_date :  
        
        for indicateur in list_indicateur:

            #création d'un sous data frame ne contenant que l'indicateur voulu        
            frame_work['bool'] = frame_work['Indicator Name'].apply(lambda indic_name: indicateur in indic_name) 
            masque_scor=frame_work['bool'] == True
            frame_indicator=frame_work[masque_scor]
            frame_indicator=frame_indicator.drop(['bool'],axis=1)          
            # normaliser la valeur des pays sur la valeur max parmis tous les pays 
            frame_normal = frame_indicator.loc[:,'2000':'2015'].apply(lambda x: x/x.max(), axis=0)  
            #frame regrouopant les indicateurs à une date donnée pour tous les pays
            frame_total[indicateur]=frame_normal.loc[:,date]
           
        #Calcul du score total
        frame_total_med=frame_total.copy()
        frame_total_med['score total']= frame_total_med.sum(axis=1) 
        #contrainte par la valeur médiane      
        median1=frame_total_med[indic_med1].median(axis=0)
        median2=frame_total_med[indic_med2].median(axis=0)
        #renvoie la valeur 0 comme score pour discrimer le pays à la date donnée   
        for i in frame_total_med.index:
            if frame_total_med.loc[i,indic_med1] <= median1:
                frame_total_med.loc[i,'score total']=0
            if frame_total_med.loc[i,indic_med2] <= median2:
                frame_total_med.loc[i,'score total']=0
                
        #frame regroupant les indicateurs à une date donnée pour tous les pays
        frame_total1[date]=frame_total_med['score total'].copy()
        frame_total_med=frame_total_med.drop('score total',axis=1)
   
    #normalisation du score total sur la valeur max de toutes les dates
    max_date=frame_total1.max()
    frame_total_norm=frame_total1.copy()
    frame_total_norm=frame_total_norm/max_date*100  
    
    #transposé pour bon format bar race
    frame_race=frame_total_norm.T.copy()
   
    return(frame_race)
```
